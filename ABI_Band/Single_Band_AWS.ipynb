**Open as a netCDF with xarray:**

```python
nc = netCDF4.Dataset(file_sel, memory = resp.content)
```

```python
ds = xr.open_dataset(xr.backends.NetCDF4DataStore(nc))
```

**Functions to add latitude and longitude:**

```python
def calc_latlon(ds):
    # The math for this function was taken from 
    # https://makersportal.com/blog/2018/11/25/goes-r-satellite-latitude-and-longitude-grid-projection-algorithm
    x = ds.x
    y = ds.y
    goes_imager_projection = ds.goes_imager_projection
    
    x,y = np.meshgrid(x,y)
    
    r_eq = goes_imager_projection.attrs["semi_major_axis"]
    r_pol = goes_imager_projection.attrs["semi_minor_axis"]
    l_0 = goes_imager_projection.attrs["longitude_of_projection_origin"] * (np.pi/180)
    h_sat = goes_imager_projection.attrs["perspective_point_height"]
    H = r_eq + h_sat
    
    a = np.sin(x)**2 + (np.cos(x)**2 * (np.cos(y)**2 + (r_eq**2 / r_pol**2) * np.sin(y)**2))
    b = -2 * H * np.cos(x) * np.cos(y)
    c = H**2 - r_eq**2
    
    #--- Added absolute to remove error
    r_s = (-b - np.sqrt(np.absolute(b**2 - 4*a*c)))/(2*a)
    
    s_x = r_s * np.cos(x) * np.cos(y)
    s_y = -r_s * np.sin(x)
    s_z = r_s * np.cos(x) * np.sin(y)
    
    lat = np.arctan((r_eq**2 / r_pol**2) * (s_z / np.sqrt((H-s_x)**2 +s_y**2))) * (180/np.pi)
    lon = (l_0 - np.arctan(s_y / (H-s_x))) * (180/np.pi)
    
    ds = ds.assign_coords({
        "lat":(["y","x"],lat),
        "lon":(["y","x"],lon)
    })
    ds.lat.attrs["units"] = "degrees_north"
    ds.lon.attrs["units"] = "degrees_east"
    return ds



def get_xy_from_latlon(ds, lats, lons):
    lat1, lat2 = lats
    lon1, lon2 = lons

    lat = ds.lat.data
    lon = ds.lon.data
    
    x = ds.x.data
    y = ds.y.data
    
    x,y = np.meshgrid(x,y)
    
    x = x[(lat >= lat1) & (lat <= lat2) & (lon >= lon1) & (lon <= lon2)]
    y = y[(lat >= lat1) & (lat <= lat2) & (lon >= lon1) & (lon <= lon2)] 
    
    return ((min(x), max(x)), (min(y), max(y)))
```

**Limiting the view to a certain range:**

```python
min_lon = -71.5
min_lat = 38
max_lon = -65
max_lat = 42.5
lats = (min_lat, max_lat)
lons = (min_lon, max_lon)
```

**Running the lat/lon calculations:**

```python
ds_ll = calc_latlon(ds)

#---This is needed to convert the lat/lon range into the x/y dimensions
#------The dataset will only filter by the official dimensions
((x1,x2), (y1, y2)) = get_xy_from_latlon(ds_ll, lats, lons)
```

```python
subset_ds = ds_ll.sel(x=slice(x1, x2), y=slice(y2, y1))
```

**Getting the wavelength and datetime strings:**

```python
dt_str = np.datetime_as_string(ds_ll.t.values, unit='s')
wl = str(ds_ll.band_wavelength.values[0])
```

**Plotting the band image:**

```python
projection=ccrs.PlateCarree()
fig,ax=plt.subplots(1, figsize=(12,12),subplot_kw={'projection': projection})
cmap = plt.cm.Wistia_r
levels=np.linspace(np.min(subset_ds.Rad), np.max(subset_ds.Rad), 15)

c=ax.contourf(subset_ds.lon, subset_ds.lat, subset_ds.Rad, cmap=cmap, levels=levels)
clb=plt.colorbar(c, shrink=0.3, pad=0.02, ax=ax)
ax.set_title('ABI '+wl+' $\mu$m ('+dt_str+')')
clb.set_label('Radiance')

ax.add_feature(cfeature.STATES)
```

**Converting the radiance to brightness temperature:**

```python
T_b = (subset_ds.planck_fk2/(np.log((subset_ds.planck_fk1/subset_ds.Rad)+1)) - subset_ds.planck_bc1)/subset_ds.planck_bc2
```

```python
projection=ccrs.PlateCarree()
fig,ax=plt.subplots(1, figsize=(12,12),subplot_kw={'projection': projection})
cmap = plt.cm.PuBu
levels=np.linspace(np.min(T_b), np.max(T_b), 15)

c=ax.contourf(subset_ds.lon, subset_ds.lat, T_b, cmap=cmap, levels=levels)
clb=plt.colorbar(c, shrink=0.3, pad=0.02, ax=ax)
ax.set_title('ABI '+wl+' $\mu$m ('+dt_str+')')
clb.set_label('Brightness Temperature (K)')

ax.add_feature(cfeature.STATES)
```
